calculadora simples

```python
import operator

num = int(input("Digite um número: "))

operations = {
    "Soma": {"symbol": "+", "func": operator.add},
    "Subtração": {"symbol": "-", "func": operator.sub},
    "Multiplicação": {"symbol": "*", "func": operator.mul},
    "Divisão": {"symbol": "/", "func": operator.truediv}
}

for op_name, op_details in operations.items():
    print(f"\n{op_name}")
    symbol = op_details["symbol"]
    func = op_details["func"]
    for i in range(1, 11):
        try:
            result = func(num, i)
            print(f"{num} {symbol} {i} = {result}")
        except ZeroDivisionError:
            print(f"{num} {symbol} {i} = Indefinido (divisão por zero)")
```
