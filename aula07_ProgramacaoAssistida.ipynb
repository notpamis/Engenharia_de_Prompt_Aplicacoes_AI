```python
# Define uma lista de números inteiros para somar
numeros = [1, 2, 3]

# Inicializa a variável 'soma' com 0. Ela armazenará a soma dos números.
soma = 0

# Inicializa a variável de índice 'i' com 0. Usaremos 'i' para percorrer a lista.
i = 0

# Inicia um laço 'while' que continua enquanto 'i' for menor que o número de elementos na lista 'numeros'
while i < len(numeros):
  # Adiciona o elemento atual da lista (numeros[i]) à variável 'soma'
  soma += numeros[i]
  # Incrementa 'i' em 1 para mover para o próximo elemento da lista na próxima iteração
  i += 1

# Imprime a lista original 'numeros' e o valor final do índice 'i' após o laço
# O 'end="| i="' garante que a próxima impressão continue na mesma linha após a lista
print(numeros, end="| i=")
print(i)
```

```python
# Define uma lista de números inteiros
numeros = [1, 2, 3]

# Calcula a soma de todos os elementos na lista 'numeros'
# A função 'sum()' é uma maneira eficiente e concisa de fazer isso no Python.
soma = sum(numeros)

# Imprime o valor da soma calculada
print(soma)
```
