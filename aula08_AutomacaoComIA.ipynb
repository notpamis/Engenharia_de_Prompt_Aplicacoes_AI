```python
import os
import shutil

def organize_files_by_extension(directory_path):
    """
    Organiza arquivos em um diretório especificado, criando subpastas
    baseadas na extensão de cada arquivo e movendo-os para lá.

    Args:
        directory_path (str): O caminho para o diretório a ser organizado.
    """

    # Verifica se o caminho do diretório fornecido existe. Se não existir, imprime um erro e sai.
    if not os.path.isdir(directory_path):
        print(f"Erro: O diretório '{directory_path}' não existe.")
        return

    print(f"Iniciando a organização do diretório: {directory_path}")

    # Define o caminho para a pasta base onde os arquivos organizados serão armazenados.
    # Esta pasta será criada dentro do diretório original.
    organized_base_dir = os.path.join(directory_path, "Arquivos_Organizados")

    # Cria a pasta base se ela não existir. 'exist_ok=True' evita um erro se a pasta já existir.
    os.makedirs(organized_base_dir, exist_ok=True)

    # Itera sobre todos os itens (arquivos e pastas) no diretório especificado.
    for filename in os.listdir(directory_path):
        file_path = os.path.join(directory_path, filename)

        # Ignora itens que são diretórios ou que são a própria pasta de organização
        # para evitar mover pastas e o diretório recém-criado.
        if os.path.isdir(file_path) or file_path == organized_base_dir:
            continue

        # Extrai o nome base do arquivo e sua extensão.
        # Ex: 'documento.pdf' -> ('documento', '.pdf')
        _, file_extension = os.path.splitext(filename)

        # Se o arquivo não tiver uma extensão (ex: 'meuarquivo' sem '.txt'), ele é ignorado.
        if not file_extension:
            print(f"Ignorando arquivo sem extensão: {filename}")
            continue

        # Cria o nome da pasta de destino com base na extensão.
        # Remove o ponto inicial da extensão (ex: '.pdf' vira 'pdf'),
        # converte para maiúsculas e adiciona '_Files' (ex: 'PDF_Files').
        folder_name = file_extension[1:].upper() + "_Files"
        destination_folder = os.path.join(organized_base_dir, folder_name)

        # Cria a pasta de destino para a extensão específica se ela ainda não existir.
        os.makedirs(destination_folder, exist_ok=True)

        # Tenta mover o arquivo para a sua nova pasta. Utiliza um bloco try-except
        # para capturar e relatar quaisquer erros que possam ocorrer durante a movimentação.
        try:
            shutil.move(file_path, os.path.join(destination_folder, filename))
            print(f"Movido: {filename} -> {folder_name}/")
        except Exception as e:
            print(f"Erro ao mover {filename}: {e}")

    print("\nOrganização concluída!")
    print(f"Verifique a pasta '{organized_base_dir}' para seus arquivos organizados.")

# --- Bloco de Execução ---
# Este bloco será responsável por criar os arquivos de teste e chamar a função de organização.

print("\nPreparando o ambiente para a demonstração...")

# Define o diretório de teste onde os arquivos serão criados
test_directory = "/content/arquivos_baguncados"

# Remove o diretório de teste e seu conteúdo, se já existir, para garantir um início limpo
if os.path.exists(test_directory):
    print(f"Removendo diretório existente: {test_directory}")
    shutil.rmtree(test_directory)

# Cria o diretório de teste
os.makedirs(test_directory, exist_ok=True)
print(f"Diretório de teste criado em: {test_directory}")

# Define os tipos de arquivo e a quantidade por tipo
file_types = {".pdf": "conteudo_pdf", ".jpg": "conteudo_jpg", ".txt": "conteudo_txt", ".docx": "conteudo_docx"}
num_files_per_type = 10

print("Criando arquivos de teste...")
# Cria os arquivos de teste
for ext, content in file_types.items():
    for i in range(1, num_files_per_type + 1):
        filename = f"arquivo_{ext[1:]}_{i}{ext}"
        file_path = os.path.join(test_directory, filename)
        with open(file_path, "w") as f:
            f.write(content)
        # print(f"Criado: {filename}") # Comentado para não poluir a saída para 40 arquivos

print(f"Total de {len(file_types) * num_files_per_type} arquivos de teste criados no diretório: {test_directory}")
print("Conteúdo inicial do diretório de teste:")
print(os.listdir(test_directory))

print("\nExecutando a organização dos arquivos...")
# Chama a função para organizar os arquivos no diretório de teste
organize_files_by_extension(test_directory)

print("\nConteúdo final do diretório de teste após a organização:")
# Lista o conteúdo do diretório de teste para ver as novas pastas e arquivos
print(os.listdir(test_directory))

print("\nVerificando o conteúdo de uma das pastas organizadas (ex: PDF_Files):")
pdf_folder = os.path.join(test_directory, "Arquivos_Organizados", "PDF_Files")
if os.path.exists(pdf_folder):
    print(os.listdir(pdf_folder))
else:
    print("Pasta PDF_Files não encontrada.")
```

```python
# ### Script Python Completo: API de Consulta de CEP com Comentários Detalhados e Saída JSON
#
# Este script integrado oferece uma solução robusta e bem documentada para consultar informações de endereço utilizando a API pública **ViaCEP**.
# Ele foi refatorado e **extensivamente comentado** em cada linha, explicando a lógica, o uso das bibliotecas e o tratamento de erros.
# Além disso, o resultado da consulta é agora formatado em JSON para facilitar a integração com outros sistemas ou para uma apresentação padronizada.
# Ideal para aprendizado e uso prático, este código combina a definição da função de consulta e um bloco de execução de demonstração em uma única célula para facilitar a execução.

import requests # Importa a biblioteca 'requests', que é essencial para fazer requisições HTTP (solicitações a servidores web).
                # Esta biblioteca simplifica o processo de enviar e receber dados de APIs RESTful.
import json     # Importa a biblioteca 'json', que é usada para trabalhar com dados no formato JSON,
                # incluindo a conversão de dicionários Python para strings JSON e vice-versa.

def consultar_cep(cep):
    """
    Função que consulta informações de endereço detalhadas para um CEP específico utilizando a API ViaCEP.

    Esta função:
    1. Limpa o CEP fornecido, removendo caracteres não numéricos.
    2. Valida o formato do CEP (deve ter 8 dígitos).
    3. Constrói a URL da requisição para a API ViaCEP.
    4. Realiza a requisição HTTP GET.
    5. Trata diversos tipos de erros (conexão, HTTP, timeout, CEP não encontrado).
    6. **Retorna os dados do endereço em formato de dicionário Python se for bem-sucedida, ou None em caso de falha.**

    Parâmetros:
        cep (str): O número do CEP a ser pesquisado. Pode ser fornecido com ou sem formatação (ex: '01001-000', '01001.000' ou '01001000').

    Retorna:
        dict ou None:
            - Um dicionário Python com os dados do endereço (logradouro, bairro, localidade, UF, etc.)
              se a consulta for bem-sucedida e o CEP for encontrado.
            - `None` em diversas situações de erro, como CEP inválido, CEP não encontrado, problemas de conexão,
              ou erros na resposta da API. Mensagens informativas serão impressas no console nesses casos.
    """

    # --- 1. Preparação e Validação do CEP ---

    # Remove todos os caracteres que não são dígitos do CEP fornecido pelo usuário.
    # A função `filter(str.isdigit, cep)` cria um iterador que retorna apenas os caracteres que são dígitos.
    # `''.join(...)` concatena esses dígitos para formar uma string limpa.
    # Isso garante que a API receba um CEP no formato esperado (apenas 8 números), independentemente da entrada do usuário.
    # Ex: se `cep` for '01001-000', `cep_apenas_digitos` será '01001000'.
    cep_apenas_digitos = ''.join(filter(str.isdigit, cep))

    # Verifica se o CEP, após a limpeza, possui exatamente 8 dígitos.
    # Um CEP brasileiro válido é composto unicamente por 8 números.
    if len(cep_apenas_digitos) != 8:
        # Se o CEP não tiver 8 dígitos, imprime uma mensagem de erro de validação.
        print(f"[Erro de Validação] O CEP informado '{cep}' (limpo: '{cep_apenas_digitos}') é inválido. Um CEP deve conter exatamente 8 dígitos numéricos.")
        return None # Interrompe a execução da função e retorna None.

    # Constrói a URL completa para a requisição à API ViaCEP.
    # Utiliza uma f-string para inserir o CEP limpo diretamente na URL.
    # O sufixo `/json/` no final da URL indica que esperamos a resposta em formato JSON (JavaScript Object Notation).
    url_da_api = f"https://viacep.com.br/ws/{cep_apenas_digitos}/json/"

    # --- 2. Realização da Requisição HTTP e Tratamento Abrangente de Erros ---

    # O bloco `try...except` é fundamental para lidar com quaisquer problemas que possam surgir
    # durante a comunicação de rede ou no processamento da resposta da API.
    try:
        # Faz uma requisição HTTP GET para a URL da API.
        # A biblioteca `requests` envia a solicitação e armazena a resposta em um objeto `response`.
        print(f"[Info] Enviando solicitação HTTP GET para a URL: {url_da_api}")
        response = requests.get(url_da_api)

        # `response.raise_for_status()` é um método conveniente que verifica o código de status HTTP da resposta.
        # Se o status for um código de erro (4xx para erros do cliente ou 5xx para erros do servidor),
        # ele levantará automaticamente uma exceção `requests.exceptions.HTTPError`.
        # Se o status for 200 OK (sucesso), esta linha não faz nada.
        print(f"[Info] Resposta HTTP recebida com status code: {response.status_code}.")
        response.raise_for_status()

        # Se a requisição foi bem-sucedida (status 200 OK) e não houve exceções HTTP,
        # a resposta é convertida de JSON para um dicionário Python.
        # `response.json()` automaticamente decodifica o corpo da resposta JSON.
        dados_recebidos = response.json()

        # A API ViaCEP tem uma particularidade: se o CEP não for encontrado na sua base de dados,
        # ela retorna um status HTTP 200 OK, mas o corpo da resposta JSON inclui uma chave 'erro' com valor `True`.
        if 'erro' in dados_recebidos and dados_recebidos['erro']:
            print(f"[Resultado] CEP '{cep_apenas_digitos}' não encontrado na base de dados da ViaCEP. Por favor, verifique o número.")
            return None # Retorna None indicando que o CEP não foi encontrado.
        else:
            # Se o CEP for encontrado e não houver erros explícitos da API, retorna o dicionário com os dados do endereço.
            print(f"[Resultado] Dados de endereço encontrados com sucesso para o CEP '{cep_apenas_digitos}'.")
            return dados_recebidos

    # --- 3. Captura e Tratamento de Exceções Específicas do Módulo `requests` ---

    except requests.exceptions.HTTPError as erro_http:
        # Captura erros que indicam falha no lado do servidor ou cliente (ex: 404 Not Found, 500 Internal Server Error).
        # Estes erros são geralmente levantados por `response.raise_for_status()`.
        print(f"[Erro HTTP] Falha na consulta para o CEP '{cep_apenas_digitos}'. Status: {response.status_code}. Detalhes do erro: {erro_http}")
        return None
    except requests.exceptions.ConnectionError as erro_conexao:
        # Captura erros que ocorrem quando não é possível estabelecer uma conexão com o servidor da API.
        # Isso pode ser devido a: ausência de conexão com a internet, falha de DNS, ou o servidor da API estar offline.
        print(f"[Erro de Conexão] Não foi possível conectar ao servidor da ViaCEP para o CEP '{cep_apenas_digitos}'. Verifique sua conexão de rede ou a disponibilidade da API. Detalhes do erro: {erro_conexao}")
        return None
    except requests.exceptions.Timeout as erro_timeout:
        # Captura erros que ocorrem quando o servidor não responde dentro de um tempo limite pré-definido.
        # Isso pode indicar que o servidor da API está sobrecarregado ou a conexão está muito lenta.
        print(f"[Erro de Timeout] A requisição para o CEP '{cep_apenas_digitos}' excedeu o tempo limite. O servidor da API pode estar lento ou indisponível temporariamente. Detalhes do erro: {erro_timeout}")
        return None
    except requests.exceptions.RequestException as erro_geral_requisicao:
        # Esta é uma classe de exceção base para todos os erros da biblioteca `requests`.
        # Captura qualquer outro erro que possa surgir durante a requisição e que não foi tratado especificamente acima.
        print(f"[Erro de Requisição Geral] Ocorreu um erro inesperado durante a requisição HTTP para o CEP '{cep_apenas_digitos}'. Detalhes do erro: {erro_geral_requisicao}")
        return None
    except Exception as erro_desconhecido:
        # Captura qualquer outra exceção Python genérica que possa ocorrer e não seja específica da biblioteca `requests`.
        # É um "catch-all" para garantir que nenhum erro inesperado passe despercebido.
        print(f"[Erro Desconhecido] Ocorreu um erro inesperado ao processar a consulta para o CEP '{cep_apenas_digitos}'. Detalhes do erro: {erro_desconhecido}")
        return None

def formatar_endereco_json(dados_endereco):
    """
    Formata um dicionário de dados de endereço em uma string JSON indentada.

    Parâmetros:
        dados_endereco (dict): Um dicionário contendo as informações do endereço.

    Retorna:
        str: Uma string JSON formatada e indentada, ou uma string JSON de erro se os dados forem None.
    """
    if dados_endereco:
        # json.dumps() converte um dicionário Python em uma string JSON.
        # `indent=4` torna a saída JSON mais legível, adicionando quebras de linha e indentação.
        return json.dumps(dados_endereco, indent=4, ensure_ascii=False)
    else:
        # Retorna uma mensagem de erro em formato JSON se não houver dados de endereço.
        return json.dumps({"erro": "Não foi possível obter dados de endereço válidos."}, indent=4, ensure_ascii=False)

# --- Bloco de Execução Principal do Script: Demonstração da Consulta de CEP com Saída JSON ---

# Este bloco de código abaixo serve como um exemplo prático de como utilizar a função `consultar_cep`
# e a nova função `formatar_endereco_json`.
# Ele guia o usuário desde a entrada do CEP até a exibição do resultado formatado em JSON.

print("\n--- Início da Demonstração da Consulta de CEP em JSON ---")

# 1. Solicita que o usuário insira um CEP.
#    A função `input()` exibe a mensagem na tela e aguarda que o usuário digite um valor e pressione Enter.
#    O valor digitado é então armazenado na variável `cep_do_usuario`.
#    Recomendações para teste:
#    - CEP válido (ex: '01001-000', '20040-009', '88010-000')
#    - CEP inexistente (ex: '99999-999')
#    - CEP com formato inválido (ex: '123', 'abcde-fg', '01001-00')
#    - Deixe vazio (pressione Enter sem digitar nada)
cep_do_usuario = input("\nPor favor, digite o CEP que deseja consultar (ex: 01001000 ou 01001-000): ")

# Imprime uma mensagem informativa para o usuário, mostrando qual CEP será consultado.
print(f"\n[Execução] O CEP fornecido pelo usuário para consulta é: '{cep_do_usuario}'")

# 2. Chama a função `consultar_cep` com o CEP que o usuário forneceu.
#    O resultado dessa chamada (um dicionário com dados ou None) é armazenado na variável `dados_endereco`.
print("[Execução] Iniciando a chamada da função 'consultar_cep' para processar o CEP...")
dados_endereco = consultar_cep(cep_do_usuario)

# 3. Formata e exibe o resultado da consulta em JSON.
#    A função `formatar_endereco_json` é chamada com os dados do endereço.
#    O resultado (uma string JSON) é então impresso.
print("\n--- [Resultado da API em JSON] ---")
json_formatado = formatar_endereco_json(dados_endereco)
print(json_formatado)

print("\n--- Fim da Demonstração da Consulta de CEP em JSON ---")
print("[Execução] Script de consulta de CEP finalizado. Obrigado por utilizar!")
```
